```python
import pandas as pd


df = pd.read_csv(
  'dasdasdasd.csv',
  usecols=['Etiqueta','Destino', 'Rampa','Data Cubagem', 'Portal Dimensionador', 'Rejeitado'],
)


df['Data Cubagem'] = pd.to_datetime(df['Data Cubagem'], format='%d/%m/%Y %H:%M:%S')
df = df.sort_values(by='Data Cubagem', ascending=False)
cod = '57J0381055'
box = df.loc[df['Etiqueta'] == cod]
boxSRT = box['Portal Dimensionador']



df = df[df['Portal Dimensionador'].isin(boxSRT)]
df = df.reset_index()


indexBox = df.loc[(df['Etiqueta'] == cod)].index[0]
df = df.drop(['Data Cubagem', 'Portal Dimensionador','index'], axis=1)
df = df.loc[indexBox-1:indexBox+1]

display(df)



```
